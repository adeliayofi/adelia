```
# 1. Inisialisasi list (array) kosong untuk menyimpan 10 bilangan
angka = []
jumlah_bilangan = 10

print(f"Masukkan {jumlah_bilangan} bilangan bulat satu per satu:")

# 2. Meminta pengguna memasukkan 10 bilangan
for i in range(jumlah_bilangan):
    while True:
        try:
            # Minta input dari pengguna
            input_bilangan = input(f"Bilangan ke-{i + 1}: ")

            # Konversi input menjadi bilangan bulat (integer)
            bilangan = int(input_bilangan)

            # Tambahkan bilangan ke dalam list 'angka'
            angka.append(bilangan)
            break # Keluar dari loop while jika input valid
        except ValueError:
            # Tampilkan pesan error jika input bukan bilangan bulat
            print("Input tidak valid. Harap masukkan hanya bilangan bulat.")

# 3. Hitung Jumlah Total (Sum)
#    Menggunakan fungsi bawaan Python sum()
total_sum = sum(angka)

# 4. Hitung Rata-rata (Average)
#    Rata-rata = Total Jumlah / Jumlah Bilangan
rata_rata = total_sum / len(angka)

# 5. Tampilkan hasil
print("\n--- Hasil Perhitungan ---")
print(f"List Bilangan yang dimasukkan: {angka}")
print(f"âž• Jumlah Total dari semua bilangan: {total_sum}")
print(f"ðŸ“Š Rata-rata dari bilangan-bilangan tersebut: {rata_rata:.2f}") # Mencetak rata-rata dengan 2 angka di belakang koma
```
